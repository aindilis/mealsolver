# Raw input tables

```python
nutrient_profiles_csv = pd.read_csv("test_nutrient_profiles.csv")
intake_requirements_csv = pd.read_csv('my_intake_requirements.csv')
log_csv = pd.read_csv('test_log.csv')
```

# Calculated data

```python
nutrient_profiles_melted = pd.melt(nutrient_profiles_csv, id_vars=['id','food','unit','serving size'],var_name='nutrient',value_name='amount')

nutrients = intake_requirements_csv.drop(['fdc_id','unit','min','upper_limit'],axis=1)

nutrient_profiles = pd.merge(nutrient_profiles_melted, nutrients, how='left', left_on='nutrient', right_on='name')

log_nutrients = pd.merge(log_csv, nutrient_profiles, how='left', left_on='id', right_on='id').fillna(0)
log_nutrients['amount_y'] = pd.to_numeric(log_nutrients['amount_y'],errors='coerce')
log_nutrients['nutrient_total'] = log_nutrients['servings']*log_nutrients['amount_y']

log_nutrients_by_date = log_nutrients.groupby(['date','internal_id']).sum()
log_nutrients_by_date.reset_index(inplace=True)
log_nutrients_by_date = log_nutrients_by_date[['date','internal_id','nutrient_total']]

daily_totals = pd.merge(log_nutrients_by_date, nutrients, how='left', left_on='internal_id', right_on='internal_id')[['date','nutrient_total','name']]
daily_totals['name'] = daily_totals['name'].fillna('to_drop')
daily_totals = daily_totals.pivot(index='date',columns='name')['nutrient_total'].reset_index().drop('to_drop',axis=1)
daily_totals.columns.name = None

intake_requirements = intake_requirements_csv.sort_values('name').set_index('name').T
intake_requirements.columns.name = None
```
