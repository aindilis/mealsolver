```python
import pandas as pd
```

```python
df = pd.read_csv("test_nutrient_profiles.csv")
df
```

```python
melted = pd.melt(df, id_vars=['id','food','unit','serving size'],var_name='nutrient',value_name='amount')
melted
```

```python
intake = pd.read_csv('my_intake_requirements.csv')
intake
```

```python
nutrients = intake.drop(['fdc_id','unit','min','upper_limit'],axis=1)
nutrients
```

```python
nutrient_profiles = pd.merge(melted, nutrients, how='left', left_on='nutrient', right_on='name')
nutrient_profiles
```

```python
log = pd.read_csv('test_log.csv')
log
```

```python
log_nutrients = pd.merge(log, nutrient_profiles, how='left', left_on='id', right_on='id').fillna(0)
log_nutrients
```

```python
log_nutrients['amount_y'] = pd.to_numeric(log_nutrients['amount_y'],errors='coerce')
```

```python
log_nutrients['nutrient_total'] = log_nutrients['servings']*log_nutrients['amount_y']
log_nutrients
```

```python
something = log_nutrients.groupby(['date','internal_id']).sum()
something.reset_index(inplace=True)
something
```

```python
something_else = something[['date','internal_id','nutrient_total']]
something_else
```

```python
daily_totals = pd.merge(something_else, nutrients, how='left', left_on='internal_id', right_on='internal_id')[['date','nutrient_total','name']]
daily_totals['name'] = daily_totals['name'].fillna('to_drop')
daily_totals
```

```python
daily_totals2 = daily_totals.pivot(index='date',columns='name')['nutrient_total'].reset_index().drop('to_drop',axis=1)
daily_totals2.columns.name = None
pd.set_option('display.max_columns', None)
daily_totals2
```
